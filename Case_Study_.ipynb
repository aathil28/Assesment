##1. Bank Account Manager
a. Simulate a bank account with deposit, withdrawal, and balance
enquiry operations

b. Use variables and data types to store account information

c. Implement conditional statements to check for sufficient balance
before withdrawal

d. Use functions to perform deposit, withdrawal, and balance inquiry
operations.


```python

def main():
  account_balance = 0
  status = True

  while status:
      print("\n ***************MENU*****************")
      print("1. Deposit")
      print("2. Withdrawal")
      print("3. Balance Enquiry")
      print("4. Exit")

      user_choice = input('Enter your choice:')

      if (user_choice == '1'):
        deposited_amount = int(input("Enter the amount you want to deposit:"))
        account_balance += deposited_amount
        print("Your deposited amount is ", deposited_amount)
        print("Current balance is:", account_balance)

      elif (user_choice == '2'):
        withdrawn_amount = int(input("Enter the amount to be withdrawal:"))
        if (withdrawn_amount <= account_balance):
            account_balance -= withdrawn_amount
            print("The amount you withdrawn is:", withdrawn_amount)
            print("Current balance is:", account_balance)
        else:
            print("Insufficient balance. Current balance is:", account_balance)


      elif (user_choice == '3'):
        print("Your balance is :", account_balance)
      elif (user_choice == '4'):
        status = False
        print("Exiting Bank Account Manager!!")
        break

      else:
        print("invalid choice!!")

if __name__ == '__main__':
    main()
```

##2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region

b. Use variables and data types to store weather data

c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)

d. Use functions to calculate average temperature and humidity, and print
results


```python
def main():
  temperatures = []
  humidities = []
  status = True

  while status:
    print("\n WEATHER ANALYSIS")
    print("1. Check weather and add data")
    print("2. Calculate average temperature and humidity")
    print("3. Exit")

    user_choice = input('Enter your choice:')

    if (user_choice == "1"):
      temperature = int(input("Your current temperature:"))
      humidity = int(input("Your current humidity:"))
      temperatures.append(temperature)
      humidities.append(humidity)

      if (temperature < 50 and humidity < 40):
        print("It's rainy today!!")
      elif (temperature > 50 and humidity > 40):
        print("It's sunny! Go out and play!!")
      else:
        print("It's a normal day")
    elif (user_choice == "2"):
      if temperatures and humidities:
        avg_temperature = sum(temperatures) / len(temperatures)
        avg_humidity = sum(humidities) / len(humidities)
        print("Average Temperature is :", avg_temperature)
        print("Average Humidity is :", avg_humidity )
      else:
        print("No weather data entered yet.")
    elif (user_choice == "3"):
      status = False
      print("Exiting Weather Data Analyzer!!")
      break
    else:
      print("Invalid choice!!")


if __name__ == '__main__':
    main()
```
